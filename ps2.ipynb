```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

file_path_cleaned = 'Sports_Dataset_Cleaned.csv'
cleaned_data = pd.read_csv(file_path_cleaned)

position_counts = cleaned_data['Position'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=position_counts.index, y=position_counts.values, palette='deep')
plt.xlabel('Position')
plt.ylabel('Number of Players')
plt.title('Number of Players by Position')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10, 10))
plt.pie(position_counts, labels=position_counts.index, autopct='%1.1f%%', colors=sns.color_palette('deep', 
                                                                                                   len(position_counts)))
plt.title('Distribution of Players by Position')
plt.show()
```

```python
from scipy.stats import chi2_contingency
expected_counts = [len(cleaned_data) / len(position_counts)] * len(position_counts)
chi2, p, dof, expected = chi2_contingency([position_counts.values, expected_counts])
print("CHI-SQUARE RESULT")
print(chi2, p)
```
