```python
import pandas as pd
import matplotlib.pyplot as plt
ile_path = 'Sports_Dataset_Cleaned2.csv'
data = pd.read_csv(file_path)

team_goals = data.groupby('Team')['Goals'].sum().reset_index()

op_team_info = team_goals.loc[team_goals['Goals'].idxmax()]
top_team = top_team_info['Team']
top_goals = int(top_team_info['Goals'])

print(f"The team with the highest number of goals is: {top_team} with {top_goals} goals")
```

```python
import matplotlib.pyplot as plt

team_goals_sorted = team_goals.sort_values(by='Goals', ascending=False)

plt.barh(team_goals_sorted['Team'], team_goals_sorted['Goals'], color='skyblue')
plt.xlabel('Number of Goals')
plt.ylabel('Team')
plt.title('Goals by Team')
plt.gca().invert_yaxis()
plt.show()


team_position_counts = data.groupby(['Team', 'Position']).size().unstack(fill_value=0)

team_position_counts.plot(kind='bar', stacked=True)
plt.xlabel('TEAM')
plt.ylabel('COUNT')
plt.title('Count of Players by Position and Team')
plt.legend(title='Position', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='-', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
goals_by_team_season = data.groupby(['Team', 'Season']).agg({'Goals': 'sum'}).reset_index()

pivot_table = goals_by_team_season.pivot(index='Season', columns='Team', values='Goals')

plt.figure(figsize=(12, 8))
pivot_table.plot(marker='o', figsize=(12, 8))
plt.title('Goals Scored by Each Team in Every Season')
plt.xlabel('Season')
plt.ylabel('Goals')
plt.xticks(rotation=45)
plt.legend(title='Team')
plt.grid(True)
plt.tight_layout()
plt.show()
```
